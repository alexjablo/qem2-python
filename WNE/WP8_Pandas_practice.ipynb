### Kaggle: Praca z danymi


Na dzisiejszych zajęciach powtórzymy główne etapy pracy z danymi. 
Dataset do pracy zawiera informacje o filmach / serialach z Netflix.



1. Ściągnij dane z : https://www.kaggle.com/shivamb/netflix-shows?select=netflix_titles.csv
2. Sprawdź liczbę kolumn oraz liczbę wierszy
3. Sprawdź, ile jest filmów oraz seriali
4. W którym miesiącu Netflix publikuje najwięcej contentu?
5. Ile filmów zostało opublikowanych w 2018?
6. Ile filmów i seriali ma kategorie "TV Dramas"?
7. Dla ilu filmów brakuje informacji o kraju pochodzenia?
